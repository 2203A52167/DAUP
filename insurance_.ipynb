```python
!pip install matplotlib

import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame and 'age', 'bmi' are the columns you want to plot
# Extract the data for the scatter plot
age = data['age']
bmi = data['bmi']

# Create the scatter plot
plt.scatter(age, bmi, marker='+', color='blue')  # Change marker to '+'

# Customize the plot
plt.title('Scatter Plot of Age vs. BMI')
plt.xlabel('Age')
plt.ylabel('BMI')

# Add text annotation (example)
plt.text(30, 35, 'Example Annotation', fontsize=12, color='red')

# Show the plot
plt.show()
```

```python
# 2. Now you can use 'data' in the following code:
plt.figure(figsize=(8, 6))  # Adjust figure size if needed
plt.scatter(data['age'], data['smoker'], marker='o', color='blue')
plt.title('Scatter Plot of Age vs. Smoking')
plt.xlabel('Age')
plt.ylabel('Smoker (yes/no)')
plt.show()
```

```python
# 3. Scatter plot between age (20 to 40) and smoking
filtered_data = data[(data['age'] >= 20) & (data['age'] <= 40)]
plt.figure(figsize=(8, 6))
plt.scatter(filtered_data['age'], filtered_data['smoker'], marker='o', color='green')
plt.title('Scatter Plot of Age (20-40) vs. Smoking')
plt.xlabel('Age')
plt.ylabel('Smoker (yes/no)')
plt.show()
```

```python
# 4. Expenses of smokers and non-smokers
smoker_expenses = data[data['smoker'] == 'yes']['expenses'].mean()
nonsmoker_expenses = data[data['smoker'] == 'no']['expenses'].mean()

plt.figure(figsize=(6, 4))
plt.bar(['Smoker', 'Non-smoker'], [smoker_expenses, nonsmoker_expenses], color=['red', 'blue'])
plt.title('Average Expenses of Smokers vs. Non-smokers')
plt.ylabel('Average Expenses')
plt.show()
```

```python
# 5. Number of smokers gender-wise
smoker_gender_counts = data[data['smoker'] == 'yes']['sex'].value_counts()
plt.figure(figsize=(6, 4))
plt.bar(smoker_gender_counts.index, smoker_gender_counts.values, color=['pink', 'lightblue'])
plt.title('Number of Smokers by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Smokers')
plt.show()
```

```python
# 6. Scatter plot between age and expenses
plt.figure(figsize=(8, 6))
plt.scatter(data['age'], data['expenses'], marker='o', color='purple')
plt.title('Scatter Plot of Age vs. Expenses')
plt.xlabel('Age')
plt.ylabel('Expenses')
plt.show()
```
